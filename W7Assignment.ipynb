```python
filename='movies.csv'
df = pd.read_csv(filename,index_col=0)
df = df.transform(lambda x: x.fillna(x.mean()))
df = df.apply(lambda x: (x-x.min(axis=0))/ (x.max(axis=0) - x.min(axis=0)))
df['Avg']=df.mean(axis=1)
df.loc['Avg']=df.mean(axis=0)
df
```

**The Advantage of using normalized data is that it scales the numbers down to ones that are more easily graphable and comparable. It also pulls outliers back into the picture by shrinking the scale. For example, if everyone in our movies example rated a movie as 1 except for one person, who gave it a 5, graphing this would look awkward because one value would be all the way on the other side of the scale. On a scale of 1-5 this might not be the worst thing, however, if the scale were 1-10,000,000 that chart would be extremely long only to accomadate that one outlier. Normalization makes the data easier to graph and comprehend when the values are extremely high or hard to read.**

```python
filename='movies.csv'
df = pd.read_csv(filename,index_col=0)
df = df.transform(lambda x: x.fillna(x.mean()))
df['Avg']=df.mean(axis=1)
df.loc['Avg']=df.mean(axis=0)
df = df.apply(lambda x: (x - x.mean()) / x.std())
df
```
