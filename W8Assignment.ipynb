```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set()

filename = "auto-mpg.data"
df = pd.read_csv(filename,delim_whitespace=True,header=None)
df.columns=['MPG','Cylinders','Displacement','Horsepower','Weight','Acceleration','Model_Year','Origin','Car_Name']
df.set_index('Car_Name',inplace=True)
df.replace('?',np.nan,inplace=True)
df['Horsepower'] = pd.to_numeric(df['Horsepower'])
df['Origin'] = df['Origin'].replace([1,2,3],['USA','Asia','Europe'])
df

```

```python
sns.distplot(df['Cylinders'], kde=False)
```

```python
sns.scatterplot(x=df['Horsepower'], y=df['Weight'])
```

## Question

**Is there a relationship between the amount of cylinders in a car and the Miles Per Gallon of the vehicle? Also, is there a relationship between the origin and either of these two attributes?** 

```python
sns.catplot(data=df,kind='swarm',x='Origin', y='MPG',hue='Cylinders')
```

## Answer

**It appears that the more cylinders a car has, the lower MPG it has, with exceptions.Also, Cars with higher cylinder counts and lower MPG tend to originate in the USA.**
