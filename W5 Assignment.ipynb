```python
filename = 'Weather.txt'
wdf = pd.read_csv(filename)
nwdf = wdf.loc[(wdf['year'] == 2013) & (wdf['month'] == 2) & (wdf['day'] == 12)]
nwdf.nlargest(5,'wind_speed')
```
